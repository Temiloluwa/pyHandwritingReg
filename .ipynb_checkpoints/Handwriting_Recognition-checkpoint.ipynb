## One vs All logistic Regression

### Computing cost function

```python
ones = np.ones(y.shape[0]).reshape(y.shape[0],1)
def sigmoid(theta,x):
    z = np.matmul(x,theta)
    z = np.exp(-1 * z) + 1
    return 1 / z

def costFunc(theta,y,x):
    htheta = sigmoid(theta,x)
    cost = (1*y* np.log(htheta)) + (ones-y)*np.log((ones-htheta))
    return np.sum(cost)/m * -1

#def costGrad(theta,y,x):
#    sig = sigmoid(theta,x)
#    sig = sig - y
#    sig = sig.T
#    grad = np.matmul(sig,x)
#    return grad.T/m

def costFuncReg(theta,y,x):
    lmd = 0.5
    theta1 = theta[1:]
    reg = lmd/m * 0.5 * np.sum(np.square(theta1))
    return costFunc(theta,y,x) + reg
    
def costGradReg(theta,y,x):
    lmd = 0.5
    print('yshape',y.shape)
    htheta = sigmoid(theta,x)
    htheta = (htheta - y)
#    htheta = htheta.reshape(htheta.shape[0],1)
    print('hthetashapea',htheta.shape)
    grad1 = np.matmul(x[...,:1].T,(htheta - y))/m #first column of X
    print('grad1shape',grad1.shape)
    grad2 = np.matmul(x[...,1:].T,(htheta - y))/m #all but first column of X
   
    grad2 = lmd/m * theta[1:] + grad2
    print('grad2shape',grad2.shape)
    return np.concatenate((grad1,grad2), axis = 0)   
```
